### Обработка пропусков

#### Высота потолков
Пустые значения оставлю пустыми, потому что пользователи их не указали, а вот с выбросами нужно разобраться.
Посмотрим распределение высоты потолков на гистограмме в диапазоне 1-5 м.

```python
histogrammer(df, x='ceiling_height', 
             xlabel = 'Высота потолка',
             ylabel='Количество квартир',
             title='Гистограмма высоты потолков',
             range=(0,5),
             bins=30
            )
```

При знакомстве с данными, мы заметили аномальные выбросы для высоты потолков.
Посмотрим, сколько у нас квартир с высотой потолков боле 4.5 метров (улсовно максимальная адекватная высота для типичных видов жилья).

```python
df.query('ceiling_height > 4.5').shape[0]
```

Изучив вопрос, выяснил, что бывает элитное жилье с высотой потолков вплоть до 10 метров. Но скорее всего такое жилье будет стоить очень дорого, за счёт своей уникальной планировки. Выясним, так ли это для наших выбросов.

Напишем функцию для построение диаграмм разброса, аналогичную функции для построения гистограмм.

```python
def scatterer(data, x=None, y=None, xlabel=None, ylabel=None, title=None, figsize=(12,6), alpha=None):
    plt.figure(figsize=figsize)
    plt.scatter(
        data=data,
        x=x,
        y=y,
        alpha=alpha
    )
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.title(title, fontsize=18)

scatterer(
    data=df.query('10 > ceiling_height > 4.5'), 
    x='ceiling_height',
    y='last_price',
    xlabel = 'Высота потолка',
    ylabel='Цена квартиры',
    title='Распределение цены квартир с высотой потолка от 4,5 до 10 метров',
            )
```

Получается, что есть 8 квартир дороже 30 миллионов и с высотой потолков около 5 метров. Это похоже на правду, а остальные выбросы в столбце `ceiling_height` заменю на NaN.

```python
df.loc[(df['ceiling_height'] > 4.5) & (df['last_price'] < 30000000), 'ceiling_height'] = np.nan
```

```python
df.loc[df['ceiling_height'] < 2, 'ceiling_height'] = np.nan
```

#### Является ли апартаментами
Тут пропущенные значения заполним `False`, так как скорее всего пользователи не заполнили этот пункт именно для обычных квартир.

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
df['is_apartment'].isna().sum() # проверка
```

#### Количество балконов
Считаем, что если значение пропущено, значит балконов нет, поэтому заполним пропуски нулями.

```python
df['balcony'] = df['balcony'].fillna(0)
df['balcony'].isna().sum() # проверка
```

#### Название населенного пункта
* Столбцы с признаками `airports_nearest` и `cityCenters_nearest` заполнены значениями только для объявлений квартир Санкт-Петербурга и населенных пунктов, включенных в его состав.
* Основываясь на этом наблюдении, заполним пропуски в `locality_name` с непустым столбцом `airports_nearest` строкой `'Санкт-Петербург'`. 
* Несмотря на эту хитрость формально мы остаёмся правы, так как пользователь точно также мог вместо пос.Парголово указать просто обобщенное название - Санкт-Петербург и весьма вероятно, что такие наблюдения на самом деле есть в наших данных.

```python
print('Было пропусков в столбце:', df['locality_name'].isna().sum())
df.loc[(df['airports_nearest'].notna())&(df['locality_name'].isna()), 'locality_name'] = 'Санкт-Петербург'
print('Оставшиеся пропуски, принадлежащие населенным пунктам Ленинградской области: ',df['locality_name'].isna().sum())
```

```python
df['locality_name'].fillna('Unknown', inplace=True)
```

Посмотрим наличие неявных дубликатов в названиях населенных пунктов.

```python
df['locality_name'].unique()
```

Смущает, конечно, что некоторые населенные пункты записаны разным образом: с приставкой "поселок"/"деревня" и без нее, с буквой "ё" или буквой "е" в слове "поселок".

Думаю, проще всего будет просто удалить тип населенного пункта и оставить только его название.

```python
remove_words = ["поселок", "посёлок", "городского", "городской", "типа", "село", "деревня", "товарищество", "коттеджный"]
remove_condition = '|'.join(remove_words)
df['locality_name'] = (df['locality_name']
                       .str.lower()
                       .replace(remove_condition,'', regex=True)
                       .apply(lambda x: x.strip())
)
print('Количество уникальных населенных пунктов ПОСЛЕ преобразования:', len(df['locality_name'].unique()),'(было 365)')
```

Добавим столбцы с днем недели, месяцем и годом публикации объявления

Добавим категорию для этажа квартиры; варианты — первый, последний, другой;

```python
def floor_category(row):
    floors = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == floors:
        return 'последний'
    else:
        return 'другой'

df.insert(
    loc=13, 
    column='floor_category', 
    value=df.apply(floor_category, axis=1))
df.head()
```

```python
df.insert(
    loc=18, 
    column='ratio_living_total', 
    value=(round(df['living_area'] / df['total_area'],2)))
df.insert(
    loc=19, 
    column='ratio_kitchen_total', 
    value=(round(df['kitchen_area'] / df['total_area'],2)))
df.head()
```

    * Наиболее дорогие квартиры располагаются в радиусе примерно 5-7 км от центра: их стоимость начинается с 5 млн и доходит до 15 млн и выше
    * **Коэффициент корреляции** хоть и небольшой, но отрицательный (-1/3), что говорит об отрицательной связи: **чем дальше от центра, тем ниже цена**.
    
**Зависимость цены квартиры от её этажа: первый, последний или другой**
* Ожидаемо, самые недорогие квартиры на 1-м этаже.
* Квартиры другого этажа чуть-чуть дороже квартир последнего этажа.
* Квартиры другого этажа также имеют больше выбросов в области высокой цены (плотный хвост).
* Меньше всего квартир с высокими ценами имеют объявления о квартирах первого этажа.
* Коэффициент корреляции с категориальной переменной посчитать невозможно.

**Зависимость цены квартиры от дня публикации**
* День недели:
    * Наиболее **высокие цены** встречаются в объявлениях, опубликованных **во вторник** (медиана 94 тыс. руб./$м^2$).
    * Разброс цен внутри каждого дня недели примерно одинаковый.
    * **Минимальные цены** на квартиры **в воскресенье** (медиана медиана 92 тыс. руб./$м^2$).
    * В каждый день недели есть публикации очень дорогих квартир, но наибольшее их количество (плотный хвост) во вторник и среду.
* День месяца: 
    * И по среднему, и по медиане **самые низкие цены на продажу квартир были в *мае-июне*, а самые высокие в *апреле***.
    * При этом наблюдается некая волнообразная сезонность: весной высокий спрос на недвижимость, летом снижается и к августу снова возрастает.
* Год:
    * В нашем датасете представлены объявления с конца 2014 по середину 2019 года, для правильного сравнения цен мы взяли диапазон длинной 5 лет с 1 мая 2015 по 1 мая 2019 года.
