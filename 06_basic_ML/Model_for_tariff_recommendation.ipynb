## Разделение данных на выборки
Для этого будем использовать функцию `train_test_split`. 
* Разделим весь датасет на обучающую и тестовую выборки в соотношении 3 к 1; 
* Для обучения будем использовать кросс-валидацию, поэтому отдельной сущностью валидационная выборка нам не понадобится.
* Для воспроизводимости выборок и модели будем использовать `randome_state=17`

```python
df_train, df_test = train_test_split(df, test_size=0.25, random_state=17)
```

```python
df_train.shape, df_test.shape
```

Теперь обучающую и тестовую выборку разделим еще на набор признаков для обучения и целевой признак.

```python
X_train = df_train.drop('is_ultra', axis=1)
y_train = df_train['is_ultra']

X_test =  df_test.drop('is_ultra', axis=1)
y_test = df_test['is_ultra']
```

## Исследование качества предсказаний разных моделей

### Дерево  решений

В проведении кросс-валидации нам поможет функция `GridSearchCV`. 
* Для поиска лучших гиперпараметров будем передавать этой функции словарь с теми значениями, которые хотим перебрать в процессе обучения.
* Будем использовать 5-кратную кросс-валидацию *(то есть, обучающая выборка будет разбита на 5 частей и цикл обучение-валидация будет повторен 5 раз, в каждой из попыток в качестве валидационной выборке будут разные части исходной выборки, но размером 1/5 от исходной выборки).*

Максимальную глубину дерева будем исследовать в диапазоне от 2 до 10.

```python
# необученное дерево
tree_grid = DecisionTreeClassifier(random_state=17)

# гиперпараметры для обучения
# можно взять еще `max_features`, но у нас итак мало признаков
tree_params = {'max_depth': range(2,11)}

# дерево, готовое стать лучшим; 
# n_jobs=-1 позволяет использовать все ядра процессора в обучении
best_tree = GridSearchCV(tree_grid, tree_params, n_jobs=-1, cv=5)    
```

```python
%%time
# обучаем лучшее дерево
best_tree.fit(X_train, y_train)
```

```python
# взглянем на параметры и долю правильных ответов у лучшего дерева
print('Лучшие гиперпараметры:', best_tree.best_params_)
print('Лучшее значение accuracy:', best_tree.best_score_)
```

**Настройка модели с лучшими гиперпараметрами**
* Теперь мы знаем, с какой глубиной дерева качество обучения наилучшее;
* Обучим с этим параметром новое дерево и сохраним как `tuned_tree`.

```python
tuned_tree = DecisionTreeClassifier(random_state=17, max_depth=7)
tuned_tree.fit(X_train, y_train)
```

### Случайный лес

Далее обучение и подбор гиперпараметров будет происходить по аналогии с деревом решений, поэтому комментарии в коде опустим.

В словарь с гиперпараметрами передаём дополнительно максимальное используемое на каждом разбиении число признаков (`max_features`).

```python
rf_grid = RandomForestClassifier(random_state=17)

forest_params = {'max_depth': range(1, 21),
                 'n_estimators': range(5,100,5)}
```

```python
best_forest = GridSearchCV(rf_grid, forest_params, cv=5, n_jobs=-1)
```

```python
%%time
best_forest.fit(X_train, y_train)
```

```python
print("Лучшие гиперпараметры для случайного леса:", best_forest.best_params_)
print("Лучшее значение accuracy", best_forest.best_score_)
```

**Настройка модели с лучшими гиперпараметрами**
* Теперь мы знаем, какая глубина одного дерева и какое количество деревьев даёт лучший результат для случайного леса;
* Обучим новый случайный лес и сохраним как `tuned_rf`.

```python
tuned_rf = RandomForestClassifier(n_estimators=85, random_state=17, max_depth=10)
tuned_rf.fit(X_train, y_train)
```

### Метод k-ближайших соседей (kNN)

* Веса (`weights`) разных соседей, то есть наших 4-х признаков, будем считать равными по вкладу в целевой признак, поэтому значение этого аргумента оставляем по умолчанию (`uniform`).

*Было бы хорошо еще нормализовать или стандартизировать каждый признак, чтобы их влияние на алгоритм обучения было одинаковым, но пока нет такой возможности :)*

И, конечно, проведем кросс-валидацию, оценив работу модели с разным количеством соседей (от 1 до 5).

```python
knn = KNeighborsClassifier()
```

```python
knn_params = {"n_neighbors": range(1, 5)}
knn_grid = GridSearchCV(knn, knn_params, cv=5, n_jobs=-1)
```

```python
%%time
knn_grid.fit(X_train, y_train)
```

```python
print("Лучшее количество соседей для метода kNN:", knn_grid.best_params_)
print("Лучшее значение accuracy", knn_grid.best_score_)
```

### Логистическая регрессия

```python
logit = LogisticRegression(random_state=17)
```

```python
%%time
logit.fit(X_train, y_train)
```

## Проверка качества моделей на тестовой выборке

**Дерево решений**

```python
tuned_tree_predictions = tuned_tree.predict(X_test)
accuracy_score(y_test, tuned_tree_predictions)
```

**Случайный лес**

```python
tuned_forest_predictions = tuned_rf.predict(X_test)
accuracy_score(y_test, tuned_forest_predictions)
```

**Метод kNN**

```python
accuracy_score(y_test, knn_grid.predict(X_test))
```

**Логистическая регрессия**

```python
accuracy_score(y_test, logit.predict(X_test))
```

### Вывод

Лучший результат показала модель случайного леса, значение accuracy = 0.813
Лучшие гиперпараметры, с которыми мы получили такой результат: 
* Количество деревьев 85
* Максимальная глубина одного дерева 10

## Проверка модели на вменяемость

Проверять адекватность модели будем с помощью классификатора `DummyClassifier`, который дает предсказания никак не опираясь на обучающие признаки. Аргумент `strategy` не меняем, по умолчанию будет стратегия, которая возвращает наиболее часто встречающийся целевой признак. 

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier()
dummy.fit(X_train, y_train)
```

```python
accuracy_score(y_test, dummy.predict(X_test))
```

### Вывод

Не может не радовать, что наша лучшая модель предсказывает всё-таки лучше, чем случайным образом.

✅Тест модели на вменяемость можно считать пройденным.
