**Переименование столбцов**

Используя пару циклов и функцию из билиотеки регулярных выражений выполним переименование столбцов из верблюжьего в змеиный регистр.

Последовательность действий цикла с двумя вложенными:
1. **Верхний цикл** просто перебирает названия всех столбцов датафрейма, название каждого столбца складываем в 2 переменные: одна хранит его как строку, вторая как список. Последнее действие цикла - добавить новую строку, приведенную к нижнему регистру, в список с новыми столбцами.
2. **Первый вложенный цикл**: в ***строке*** с названием колонки функцией `finditer` выполняем поиск по паттерну "большие латинские буквы", индекс каждого вхождения большой буквы складываем в список `points_of_start`.
3. **Второй вложенный цикл**: здесь мы перебираем список с индексами вхождения больших букв, *отбросив первую большую букву*,  работаем с названием колонки, преобразованной в ***список букв***, вставляя после каждой большой буквы символ "_".
4. Последним действием, уже вне циклов, присваиваем список с новыми названиями переменной `columns`, принадлежащей объекту `DataFrame`.

```python
for name in df.columns:
    rena = re.sub(r"(?=[A-Z])(?!^)", '_', name).lower()
    df = df.rename(columns={name:rena})
```

## Обучение моделей без учета дисбаланса классов

**Функция для проведения обучения**

Для удобства итерирования с помощью кросс-валидации на разных моделях создадим небольшую функцию, принимающую на вход:
* модель;
* словарь с гиперпараметрами;
* обучающую выборку и целевой признак;
* метрику(и) для оценки моделей в процессе кросс-валидации; 
* число разбиений обучающей выборки для кросс-валидации.

Возвращает функция список, содержащий значения двух метрик: F1 и ROC-AUC.

```python
def best_model_by_cv(model, params, features_train=features_train, target_train=target_train, 
                     scoring=['f1', 'roc_auc'], cv=5):
    model_grid = GridSearchCV(model, params, scoring=scoring, n_jobs=-1, cv=cv, refit='f1')
    model_grid.fit(features_train, target_train)
    
    results = [model_grid.cv_results_.get("mean_test_f1").max(), 
               model_grid.cv_results_.get("mean_test_roc_auc").max()]
    
    print(f'Лучшие гиперпараметры модели: {model_grid.best_params_}')
    print(f'Лучший показатель метрики F1: {results[0]}')
    print(f'Лучший показатель метрики ROC-AUC: {results[1]}')
    return results
```

Список с лучшими показателями каждой модели будем записывать в датафрейм, чтобы потом сравнить разные модели на выборках с разными условиями обучения.

    * Постфикс `_unbal` означает, что обучение шло на несбалансированной выборке; 
    * Постфикс `_up` означает, что обучающая выборка была искуствено увеличена;

### Upsampling

Посмотрим, какого размера наша выборка до балансировки классов.

```python
features_train.shape, target_train.shape
```

Функция для балансировки классов в обучающей выборке путём умножения позитивных наблюдений до количества, чтобы баланс классов стал 1:1.

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=random_state)
    
    return features_upsampled, target_upsampled
```

### Downsampling

Функция для балансировки классов в обучающей выборке путём случайного удаления негативных наблюдений до количества, чтобы баланс был 1:1.

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=random_state)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=random_state)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=random_state)
    
    return features_downsampled, target_downsampled
```

```python
features_downsampled, target_downsampled = downsample(features_train, target_train, 0.2)
features_downsampled.shape, target_downsampled.shape
```
