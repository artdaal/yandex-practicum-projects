## Обучение модели

Функция `split_and_fit`:
* разделяет данные на обучающую и валидационную выборки, 
* обучает линейную регрессию 
* выводит на экран значение метрики RMSE на валидационной выборке
* возвращает обученную модель

```python
results = pd.DataFrame(index=['RMSE', 'mean_predicted_volume', 'real_values', 'predict_values'])                
```

```python
def split_and_fit(data, test_size=0.25):
    X = data.drop(['id', 'product'], axis=1)
    y = data['product']
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, 
                                                        random_state=random_state)

    model = LinearRegression()
    model.fit(X_train, y_train)
    predictions =  model.predict(X_test)
    rmse = mean_squared_error(y_test, predictions)**0.5    
    return rmse, predictions.mean(), y_test, predictions 
```

```python
results['region_1'] = split_and_fit(df1)
```

```python
results['region_2'] = split_and_fit(df2)
```

```python
results['region_3'] = split_and_fit(df3)
```

## Расчет прибыли
### Расчетная прибыль для безубыточного уровня

Вводные данные:

* `locations_to_research` - количество скважин при разведке нефти в регионе;
* `locations_to_work` - количество скважин, выбираемых для добычи;
* `budget` - бюджет на разработку всех скважин в регионе;
* `price_per_barrel` - цена одного барреля нефти;
* `profit_per_unit` - доход с каждой единицы продукта (тысячи баррелей);
* `mvp` *(minimum viable product)* - минимальный средний объём добываемого сырья из одной скважины для выхода региона в безубыток.

```python
locations_to_research = 500
locations_to_work = 200
budget = 10e9
price_per_barrel = 450
profit_per_unit = 450e3
```

```python
mvp = budget / (profit_per_unit * locations_to_work)

print(f'Объем нефти, необходимый для безубыточной разработки = {mvp:.3f} (тысяч баррелей)')
```

### Расчет прибыли по предсказаниям добычи из скважин

Предсказанные и реальные значения для тестовой выборки, у нас хранятся в датафрейме `results`.

Далее рассчитаем для 200 наиболее перспективных скважин каждого региона:
* Суммарный объём предполагаемого сырья
* Прибыль для этого объёма сырья

Функция принимает для подсчёта прибыли принимает на вход:
* `pd.Series` с предсказаниями по региону,
* `pd.Series` с реальными объёмами по региону.

```python
def profit_counter(predict_values, real_values, random_state=random_state, output=True) -> float:
    
    # здесь берем выборку из предсказаний модели, сортируем по возрастанию значений, берём индексы 200 лучших
    indexes_from_sample = predict_values.sample(n=locations_to_research, random_state=random_state)\
                                .sort_values(ascending=False).head(locations_to_work).index
    
    # чтобы индексация прошла правильно сбросим индексы у реальных значений
    real_values = real_values.reset_index(drop=True)
    
    # рассчитываем показатели по реальным данным, но на основе индексов из предсказаний
    mean_supply = real_values.loc[indexes_from_sample].mean()
    sum_supply = real_values.loc[indexes_from_sample].sum()
    profit_supply = (real_values.loc[indexes_from_sample].sum() * profit_per_unit - budget) / 1e6
    
    if output:
        print(f'Средний запас сырья в выборке из 200 лучших скважин {mean_supply:.2f} тысяч баррелей')
        print(f'Суммарный запас сырья в выборке из 200 лучших скважин {sum_supply:.2f} тысяч баррелей')
        print(f'Суммарная прибыль при реализации всего запаса сырья из 200 лучших скважин {profit_supply:.2f} млн. руб.\n')
        return
    return profit_supply
```

```python
for region in (results.loc['predict_values',:].index):
    print(region.upper().replace('_', ' '))
    
    predict_values = pd.Series(results.loc['predict_values', region])
    real_values =  results.loc['real_values', region]
    
    profit_counter(predict_values, real_values)
```

#### Вывод

Теперь акценты немного сместились:
* Средний запас сырья в одной скважине наибольший в третьем регионе - 118,42 тысяч баррелей.
* Так как число скважин для планируемой разработки одинаково (200), то при бОльшем среднем, бОльший суммарный объём также в третьем регионе.
* Поэтому **самым выгодным является третий регион**, где суммарная прибыль получилась 657 млн.руб.

## Расчет прибыли с учетом рисков

```python
%%time
state = np.random.RandomState(42)
bootstrap_samples = 1000


for region in (['region_1', 'region_2', 'region_3']):
    
    predict_values = pd.Series(results.loc['predict_values', region])
    real_values =  results.loc['real_values', region]
    region = region.upper().replace("_", " ")
    profit_sum = pd.Series()
    
    for i in range(bootstrap_samples):
                
        profit = pd.Series(profit_counter(predict_values, real_values, random_state=state, output=False))
        profit_sum = pd.concat([profit_sum, profit])
        
        ci_down = profit_sum.quantile(0.025)
        ci_up = profit_sum.quantile(0.975)
        risk = (profit_sum < 0).mean()
    print(f'{region}')
    print(f'Средняя прибыль (Bootstrap, 1000): {profit_sum.mean():.2f} млн.руб')
    print(f'Доверительный интервал для прибыли: от {ci_down:.2f} до {ci_up:.2f} млн.руб')
    print(f'Риск уйти в  убыток: {risk:.2%}')
    print(f'ROI {profit_sum.mean()/(budget/1e6):.2%} \n')
```
